```python
import pandas as pd

df2 = pd.read_sql_table('wroclaw_data', 'sqlite:///wroclaw_nuremberg_public_transport.sqlite')
```

```python
df2.head(10)
```

```python
df2.info()

df2['Betriebszweig'].unique()
```

```python
import matplotlib.pyplot as plt
sums = df.groupby(df["Betriebszweig"])["latitude"].sum()
plt.axis('equal');
plt.pie(sums,autopct='%1.0f%%', labels=sums.index);
plt.show()
```

```python
import matplotlib.pyplot as plt2
sums2 = df2.groupby(df2["Betriebszweig"])["latitude"].sum()
plt2.axis('equal');
plt2.pie(sums2,autopct='%1.0f%%', labels=sums2.index);
plt2.show()
```

```python
import plotly.express as px
import plotly.express as px2
import pandas as pd
color_scale = [(0, 'purple'), (1,'green')]
fig = px.scatter_mapbox(df, 
                        lat="latitude", 
                        lon="longitude", 
                        hover_name="Betriebszweig", 
                        hover_data=["Betriebszweig"],
                        color="Betriebszweig",
                        color_continuous_scale=color_scale,
                        
                        zoom=9, 
                        height=500,
                        width=500)
fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
color_scale = [(0, 'purple'), (1,'green')]
fig2 = px2.scatter_mapbox(df2, 
                        lat="latitude", 
                        lon="longitude", 
                        hover_name="Betriebszweig", 
                        hover_data=["Betriebszweig"],
                        color="Betriebszweig",
                        color_continuous_scale=color_scale,
                        zoom=9, 
                        height=500,
                        width=500)
fig2.update_layout(mapbox_style="open-street-map")
fig2.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig2.show()
```
